```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load the blood sugar data
data = pd.read_csv('diabetes_data.csv', index_col=0, parse_dates=True)

# Split the data into training and test sets
train_data = data.iloc[:800]
test_data = data.iloc[800:]

# Fit a SARIMAX model on the training data
model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(0, 1, 1, 12))
results = model.fit()

# Generate forecasts for the test data
forecasts = results.forecast(steps=200)

# Plot the historical data and the forecasted data
plt.figure(figsize=(12, 6))
plt.plot(train_data, label='Historical Data')
plt.plot(test_data, label='True Future')
plt.plot(forecasts, label='Forecasted Data')
plt.legend()
plt.show()
```
